## 8. Plot the cumulative Frequency Polygon for "Total Bill" and "Tip"

### CUMULATIVE FREQUENCY POLYGON FOR "TOTAL BILL"

```python
v1, b1 = np.histogram(bill)
cumulative = np.cumsum(v1)      
plt.plot(b1[:-1], cumulative)
plt.title('Total Bill',fontsize=15)
plt.show()
```

### CUMULATIVE FREQUENCY POLYGON FOR "TIP"

```python
v2, b2 = np.histogram(tip, bins=40)
cumulative = np.cumsum(v2)
plt.plot(b2[:-1], cumulative)
plt.title('Tip',fontsize=15)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```

## 10. Prepare two variable frequency tables for "Size" vs "Tip" and "Size" vs "Total Bill". Summarize and explain your findings.

### SIZE VS TIP

```python
tipSize = pd.crosstab(index=imp["size"],columns=imp['tip'], margins=True)

tipSize.index=["1 member","2 members","3 members","4 members","5 members","6 members","Total"]

print(tipSize)
print("\n Through this table I can found that how much tip we recieve from families with different number of peoples.")
```

### SIZE VS TOTAL BILL

```python
billSize = pd.crosstab(index=imp['size'],columns=imp['total_bill'],margins=True)
billSize.index=["1 member","2 members","3 members","4 members","5 members","6 members","Total"]
print(billSize)
print("\n Through this table I can easily found that how much worth of food did "
      "families buy families with different number of peoples.")
```
